## CONCATENAR Dataframes >> ``pd.concat([df1,df2,..,dfn],axis=1)``
- **``axis=1``** es importante porque va a poner una columna al lado de la otra


- **``pd.concat``** solamente copia todas las columnas de un DataFrame y de otro sin chequear si tienen el mismo nombre, por ejemplo.**NO reconoce columnas con el mismo nombre**. Si queremos mergear 2 DataFrames con algunas columnas que se llamen igual y que esto  lo reconozca debemos utilizar **``pd.merge``**.


- Si los DataFrames no tienen la **misma cantidad de filas**, rellena con NaN las vacías.

```python
abcdef = pd.concat([abcd,ef],axis=1)
abcdef.head()
```

```python
#  Concatenar 1 DATAFRAME y 1 sola columna de otro DataFrame 
abcdf = pd.concat([abcd,ef['F']],axis=1)
abcdf.head()
```

## MERGE Dataframes >> ``pd.merge(df1,df2,...,dfn,on='col')``
- **``pd.merge``** junta DataFrames que tengan **1 ó más columnas en comun**. Va a utilizar la columna **``on='col'``** como fija, y va a copiar todo el resto de columnas una al lado de la otra, dejando la **``'col'``** intacta.


```python
t1 = [[1,'alice@gmail.com',20],[2,'bob@hotmail.com',25],[3,'carol@yahoo.com',30]]
t2=[[1,1,1],[1,2,0],[1,5,0],[2,3,0],[2,4,1],[2,1,0],[3,2,0],[3,1,0],[3,3,0],[3,4,0],[3,5,1]]

t1_df = pd.DataFrame(t1,columns=['user_id','email','age'])
t2_df = pd.DataFrame(t2,columns=['user_id','add_id','click'])

tdf = pd.merge(t1_df,t2_df,on='user_id')
tdf
```

## Apply DEF a columna >> ``df['col'].apply(DEF_NAME)``

```python
# Aplico la DEF solamente a la columna "r1"

def raiz(x):
    return x**0.5

my_array_df['r1raiz'] = my_array_df[['r1']].apply(raiz)
my_array_df.head()
```
