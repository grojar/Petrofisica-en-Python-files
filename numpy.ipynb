##  np.matmul(a,b) >> DOT Product  Producto escalar de vectores
https://matematicasconmuchotruco.wordpress.com/2014/09/10/producto-escalar-en-el-plano-y-aplicaciones/<br>


- Para poder hacer el producto escalar de 2 matrices, necesito que **n columnas de M1 = n filas de M2**, lo que se conoce como **inner dimension**.


- Si el resultado es una matriz , las dimensiones de la **matriz resultante = (n filas M1, n columnas M2)**.



```python
a = np.array([1,2])
b = np.array([3,4])
np.matmul(a,b)
```

```python
# size "a"=(2,3) ... size "b"=(3,4) ...(col "a"=3...fila "b"=3) PODEMOS MULTIPLICAR ESCALARMENTE
# La matriz resultante es: (2,4), tiene la cantidad de filas de "a", y la cantidad de columnas de "b"
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[2,4,5,2],[3,1,2,5],[2,1,3,2]])
np.matmul(a,b)
```

```python
a = np.array([[1,4],[4,5]])
b = np.array([[2,2],[3,3]])
np.matmul(a,b)
```

## RANDOM Normal/Gaussian >> ``np.random.randn(fil,col)``
- Me devuelve una matriz de dimensiones **``(fil,col)``** con valores entre **0-1** y con una **distribucin normal o gaussiana**.

```python
np.random.randn(2,3)
```

## RANDOM Normal/Gaussian >> ``np.random.normal(loc=n,scale=n,size=(n,n))``
- Me devuelve una matriz donde los datos tienen una distribucion **Gaussiana o Normal** de dimensiones **``size=(n,n)``** con valores **centrados** en **``loc=n``** y con una **standar deviation** **``scale=n``**.

```python
np.random.normal(loc=3,scale=2,size=(4,3))
```

## RANDOM Uniform>> ``np.random.uniform(low=n,high=n,size=(n,n))``
- Devuelve valores aleatorios en una **Distribucion Uniforme** (Distribucin Uniforme es cuando todos los valores entre un minimo y un maximo tienen las mismas posibilidades. Un ejemplo es lanzar un dado, ya que todos los 6 resultados posibles tienen 1/6 de probabilidad de ocurrencia cada uno)
- Pueden repetirse valores (de hecho, probando aparecieron valores identicos ente si)

```python
np.random.uniform(low=0,high=45,size=(5,3))
```

## SEED >> ``np.random.seed(number)``
- Me devuelve valores random, pero siempre son los mismos. Se debe escribir esta linea **antes** que la linea de creacion de la matriz **(np.random.randint, etc)**.

```python
np.random.seed(42)              # Se deben correr las 2 lineas juntas SIEMPRE.
np.random.randint(0,100,(3,4))  # no importa cuantas veces corra esto, siempre voy a obtener la misma matriz.
```

## NOISE (agregar ruido)

```python
# 100 valores separados linealmente
x_data = np.linspace(0.0,10.0,100)

# 100 valores random
noise = np.random.randn(len(x_data))

# Hago una ecuacion lineal con y sin ruido
y_data = (0.5 * x_data ) + 5
y_noise = (0.5 * x_data ) + 5 + noise


plt.scatter(x_data,y_noise,marker='.')
plt.plot(x_data,y_data,color='r')
plt.xlabel('X',size=12)
plt.ylabel('Y',size=12)
```
